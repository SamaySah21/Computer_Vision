#### In this notebook, we are just looking into the very basic concepts of image reading such as load image, getting shape, channels attributes using color map.

```python
## Loading the basic Libraries

import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading and Reading the image

img = io.imread("jethalal.jpg")
plt.imshow(img)
plt.show()
```

```python
#getting shape of image
img.shape
```

```python
## Getting image channel attributes separately for RGB image
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]
```

```python
## Showing red color map of red channel
plt.imshow(red, cmap = "Reds")
plt.show()
```

```python
## Showing green color map of green channel
plt.imshow(green, cmap = "Greens")
plt.show()
```

```python
## Showing blue color map of blue channel
plt.imshow(blue, cmap = "Blues")
plt.show()
```

```python
# Crop Image part img(height, width)
## height = b/w 100 to 200 px will crop
## width = b/2 120 to 300 px will crop

plt.imshow(img[100:200, 120:300])
```
